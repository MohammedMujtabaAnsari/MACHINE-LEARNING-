```python
from sklearn.model_selection import train_test_split

# Identify categorical columns
categorical_cols = features.select_dtypes(include='object').columns

# Apply one-hot encoding to the categorical columns
x_encoded = pd.get_dummies(features, columns=categorical_cols, drop_first=True)

xtrain,xtest,ytrain,ytest=train_test_split(x_encoded,target,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(xtrain,ytrain)
ypred=logreg.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
ac=accuracy_score(ytest,ypred)
cm=confusion_matrix(ytest,ypred)
clr=classification_report(ytest,ypred)

print(f"Accuracy Score:{ac}\n COnfusion Matrix:{cm}\n Classifcation_Report\n{clr}")
```

```python
reviews=["I don't like a colour of my car","I need to take my car to the mechanic for a check","Less than a month after purchase,the vehicle is experiencing severe engine overheating","Upon receiving my car, i noticed that your team faiiled to rectify the engine  oil leakage"]
```

```python
lower_text=[]
for sentence in reviews:
  lower_text.append(sentence.lower())
print("LowerCase :",lower_text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
x=cv.fit_transform(reviews)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(10, input_dim=x.shape[1], activation='relu'))
model.add(Dense(5, activation='relu'))
model.add(Dense(3, activation='softmax')) # Happy / Neutral / Angry
```

```python
import numpy as np
from tensorflow.keras.utils import to_categorical

# Assuming sentiment labels: 0 for positive, 1 for neutral, 2 for negative
# "I don't like a colour of my car" -> Assuming Neutral (1) for general dissatisfaction
# "I need to take my car to the mechanic for a check" -> Assuming Neutral (1)
# "Less than a month after purchase,the vehicle is experiencing severe engine overheating" -> Negative (2)
# "Upon receiving my car, i noticed that your team failed to rectify the engine  oil leakage" -> Negative (2)
labels = np.array([1, 1, 2, 2])

# Convert integer labels to one-hot encoded format
num_classes = 3 # Based on the model's output layer
y = to_categorical(labels, num_classes=num_classes)

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x.toarray(), y, epochs=150)
```

```python
from sklearn.naive_bayes import MultinomialNB

# Re-define labels to match the number of reviews
# "I don't like a colour of my car" -> Assuming Neutral (1)
# "I need to take my car to the mechanic for a check" -> Assuming Neutral (1)
# "Less than a month after purchase,the vehicle is experiencing severe engine overheating" -> Negative (2)
# "Upon receiving my car, i noticed that your team failed to rectify the engine  oil leakage" -> Negative (2)
labels_updated = np.array([1, 1, 2, 2])

ml_model = MultinomialNB()
ml_model.fit(x, labels_updated)
prediction = ml_model.predict(cv.transform([' i did not  rwd cars']))
print(prediction)
```
