```python
# Step00: Data
```

```python
text=["Win Money Today",
      "Claim your Price",
      "Meeting at 5 pm",
      "Project submission today"]

label=[1,1,0,0] # 1=Spam,0=No Spam
```

```python
# Step 01: LowerCasing
```

```python
lower_text=[]
for sentence in text:
  lower_text.append(sentence.lower())
print("LowerCase :",lower_text)
```

```python
#Step 02: Tokenzation
token_text=[]
for sentence in lower_text:
  words=sentence.split()
  token_text.append(words)


print("Tokenization :",token_text)


# Step 03: Stopwords Removal
import nltk
nltk.download('stopwords')   # stopword is identify the words like (the,is,these, their)

from nltk.corpus import stopwords
stop_words=stopwords.words("english")


clean_text=[]
for sentence in token_text:
    new_words=[]
    for word in sentence:
      if word not in stop_words:
        new_words.append(word)
    clean_text.append(new_words)


print("Stopwords Removal:",clean_text)
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()

stem_text=[]
for sentence in clean_text:
  new_words=[]
  for word in sentence:
    root_word =ps.stem(word)
    new_words.append(root_word)
  stem_text.append(new_words)
print("Steaming:",stem_text)
```

```python
# Step 05: Join Words Again
final_text=[]
for sentence in stem_text:
  sentence=" ".join(sentence)
  final_text.append(sentence)

print("Final Clean Text:", final_text)
```

```python
# Step 06: Convert Words to Numbers (Vectorization)
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer()
x = cv.fit_transform(final_text)

print("Vectorization:\n",x.toarray())
```

```python
#Step07:Apply Naive Bayes
from sklearn.naive_bayes import MultinomialNB

model=MultinomialNB()
model.fit(x,label)

test=["Win prize now"]

test=[test[0].lower()]

test_words=[]
for word in test[0].split():
  test_words.append(ps.stem(word))
```

```python
# Step 08: Predict on the test sentence
test_vector = cv.transform([" ".join(test_words)])
prediction = model.predict(test_vector)

if prediction[0] == 1:
    print("Prediction: Spam")
else:
    print("Prediction: Not Spam")
```
