#Model Building

```python
x=df.iloc[:,:-1] #Dataframe
y=df.iloc[:,-1] #Series
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
def mymodel(model):
  #Create a model(model creation)
  model.fit(xtrain,ytrain)
  ypred=model.predict(xtest)

  #Check bais and variance
  train= model.score(xtrain,ytrain)
  test= model.score(xtest,ytest)
  print(f"Training accuracy:- {train} \n testing accuracy:- {test}")

  #model evaluation
  print(classification_report(ytest,ypred))
  return model
```

TO find the categorical data accuracy we use these MOdels

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
KNN=mymodel(KNeighborsClassifier())
```

```python
LR=mymodel(LogisticRegression())
```

```python
svc=mymodel(SVC())
```

```python
DT=mymodel(DecisionTreeClassifier())
```

#Hyper Parameter Tunning
 it is  technique to find the ideal value for specific parameter

```python
parameters= {
    "criterion":["gini","entropy"],
    "max_depth":list(range(1,20)),
    "min_samples_leaf":list(range(1,20))
}
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(DecisionTreeClassifier(),parameters,verbose=2)
grid.fit(xtrain,ytrain)
```

```python
dt=mymodel(grid.best_estimator_)
```
